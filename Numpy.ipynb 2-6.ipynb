```python
import numpy as np
```

#### combining arrays

#### Horizontal stack(hstack)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.hstack((a,b))
print(result)
```

#### vertical stack(Ustack)

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.vstack((a,b))
print(result)
```

#### spliting arrays

```python
arr = np.array([1,2,3,4,5,6])
result = np.split(arr,2)
print(result)
```

## search function

#### find index

```python
arr = np.array([10,20,30,40,50])
index = np.where(arr == 30)
print(index)
```

#### search sorted position

```python
arr = np.array([10,20,30,40,50])
position = np.searchsorted(arr,25)
print(position)
```

#### sort function

```python
arr = np.array([10,20,30,40,50])
print(np.sort(arr))
```

## Aggregation & statistical function

```python
arr = np.array((10,20,30,40,50))
print("sum:",np.sum(arr))
print("mean:",np.mean(arr))
print("median:",np.median(arr))
print("max:",np.max(arr))
print("min:",np.min(arr))
```

#### Filter arrays(Bolean indexing)

```python
arr = np.array([10,20,30,40,50])
result = arr[arr < 35]
print(result)
```

#### Multiple condition

```python
arr = np.array([10,20,30,40,50])
result = arr[(arr > 20)&(arr < 50)]
print(result)
```

#### Masking arrays

```python
arr = np.array ([10,20,30,40,50])
mask = arr > 30
print(mask)
```

```python
filtered = arr[mask]
print(filtered)
```

#### Reshaping arrays

##### convert array dimention without changing data.

```python
# 1D to 3D
arr = np.array([10,20,30,40,50,60,70,80])
result = arr.reshape(2,2,2)
print(result)
```

```python
# 1D to 2D
arr = np.array([10,20,30,40,50,60])
result = arr.reshape(2,3)
print(result)
```

```python
arr = np.array([[[10,20],[30,40]],[[50,60],[70,80]]])
result = arr.reshape(8)
print(result)
```

#### Array broadcasting

```python
a = np.array([[1],[2],[3]])
b = np.array([10,20,30])

result = a+b
print(result)
```
